# Text Generation Using Simple GPT2
Prepared to be executed in Google Colab, which has parallel processing and is well suited for training the simpleGPT2 model.

```python
%tensorflow_version 1.x
!pip install -q gpt-2-simple
import gpt_2_simple as gpt2
```

```python
!nvidia-smi
```

```python
# Download 124M GPT2 model
gpt2.download_gpt2(model_name = '124M')
```

```python
# Connect to google drive (when using Colab)
gpt2.mount_gdrive()
```

```python
file_name = "OSE_OnlyTexts.csv"
```

```python
gpt2.copy_file_from_gdrive(file_name)
```

```python
sess = gpt2.start_tf_sess()
```

```python
gpt2.finetune(sess,
              dataset=file_name,
              model_name='124M',
              steps=1000,
              restore_from='fresh',
              run_name='run1',
              print_every=10,
              sample_every=200,
              save_every=500
              )
```

```python
# Save to google drive
gpt2.copy_checkpoint_to_gdrive(run_name='run1')
```

# Text Generation
Load trained simpleGPT2 model and generate text.

```python
# Import finetuned model from 
gpt2.copy_checkpoint_from_gdrive(run_name='run1')
```

```python
sess = gpt2.start_tf_sess()
gpt2.load_gpt2(sess, run_name='run1')
```

```python
gpt2.generate(sess, run_name='run1')
```
